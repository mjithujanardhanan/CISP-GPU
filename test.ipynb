```python
from tkinter import *
from ttkbootstrap.constants import *
import ttkbootstrap as tb

```

```python
def click():
    global counter
    counter += 1
    if(counter & 1):
        my_label.config(text = " Hello World")
    else:
        my_label.config(text = "bye world")
        
```

```python
root = tb.Window( themename = "superhero");
counter = 0



root.title(" TTK BootStrap!")
#root.iconbitmap('images/codemy.ico')
root.geometry('250x175')

my_label = tb.Label(text = "Hello World!", font = ('Times New Roman', 28), bootstyle = "default")
my_label.pack()

my_button = tb.Button(text = "Click Me", bootstyle = "info", command = click)
my_button.pack()

root.mainloop()
```
